# Word Cloud of AI Generated Summaries

This code uses the Turkish Political Economy Biographical Library and generates a word cloud of the AI summaries of each row in the dataset.

# Citation: #
Yalçıntaş, A., Kızılyurt, Ö., Başaran, V., Tok, A. Voyvoda, N., Gelmedi, Ş., Bal, E., Kasap, O. (n.d.). The Turkish Political Economy Database, from the 1800s to Date. Retrieved from [Github](https://github.com/sekerefe/TRPolEcon_Public)

```python
# Import libraries for data manipulation, text processing, word cloud generation, and visualization
import pandas as pd 
import nltk 
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
from nltk.corpus import stopwords 

# Ensure NLTK stopwords are downloaded
nltk.download('stopwords', quiet=True) 

# Path to the CSV file
csv_path = r'.\files\biographical_library.csv' 
df = pd.read_csv(csv_path)

# Concatenate all AI summaries into a single string
total_text = ' '.join(df['ai_summary'].dropna().astype(str)) 
# Get English stop words from NLTK
stop_words = set(stopwords.words('english'))
# Filter out stop words from the text 
words = [word for word in total_text.split() if word.lower() not in stop_words] 
# Join the filtered words back into a single string for word cloud generation
filtered_text = ' '.join(words) 

# Generate the word cloud from the filtered text
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(filtered_text) 
# Set the figure size for the plot
plt.figure(figsize=(15, 7)) 
# Display the word cloud image
plt.imshow(wordcloud, interpolation='bilinear')
# Turn off the axis 
plt.axis('off') 
# Set the title for the plot
plt.title('Word Cloud of AI Generated Summaries') 

```
