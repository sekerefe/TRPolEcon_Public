**Nationality Distribution**

This code uses the dataset of *The Turkish Political Economy Biographical Library* in *The Turkish Political Economy Database, from the 1800s to Date*. It generates a bar chart of nationality distribution of scholars in the dateset.

**Citation**: Yalçıntaş, A., Kızılyurt, Ö., Başaran, V., Tok, A. Voyvoda, N., Gelmedi, Ş., Bal, E., Kasap, O. (n.d.). *The Turkish Political Economy Database, from the 1800s to Date*. Retrieved from [Github](https://github.com/sekerefe/TRPolEcon_Public)

```python
# Import necessary libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import os

# Read the CSV file containing biographical data
csv_path = r'C:..\files\biographical_library.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_path)

# Calculate the counts of each unique nationality
nationality_counts = df['nationality'].value_counts()

# Create and Display the Bar Chart ---
plt.style.use('seaborn-v0_8-whitegrid')
# Adjust figure size for better readability
plt.figure(figsize=(14, 7))  

# Create the bar plot
nationality_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Add titles and labels for clarity
plt.title('Distribution of Nationalities in Biographical Library', fontsize=16)
plt.xlabel('Nationality', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate labels for better fit
plt.tight_layout()  # Adjust layout to prevent labels from overlapping

# Display the plot
plt.show()
```
